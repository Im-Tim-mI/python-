```python
print('華氏溫度轉攝氏')
ff=float(input('輸入華氏溫度:'))
cc=(ff-32.0)*5.0/9.0
nn=u'\N{DEGREE SIGN}'
print(f'華氏{ff}{nn}F等於攝氏{cc}{nn}C')
```

```python
import math
print('圓周長,面積')
r=float(input('請輸入半徑:'))
PI=math.pi
ff=PI*r*r
cc=2*PI*r
print(f'半徑={r}CM 圓面積={ff}CM2 圓周長={cc}CM')
```

```python
print('BMI')
g=float(input('請輸入體重(公斤):'))
m=float(input('請輸入身高(公尺):'))
BMI=g/(m*m)
print(f'你的BMI:{BMI}')
```
