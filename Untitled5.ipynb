```python
def accumlated_amount(principal,year_rate,period=12):
  period = period
  month_rate = year_rate / 100 / 12
  month_rate_coeff = 1+ month_rate
  cofficient = month_rate_coeff * (month_rate_coeff ** period - 1) / month_rate
  amount = principal * cofficient

  return amount
```

```python
principal = int(input('Principal: '))
year_rate = float(input('Year rate(%):'))
period = input('period: ')
```
