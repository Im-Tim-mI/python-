```python
q_list = [
  '蘋果的英文?',
  '一打雞蛋用掉兩個,還剩下幾個?',
  '「問世間,情為何物?」的下一句?',
  '一天一個樣,之後又重複。(猜一天體)'
]
a_list = ['apple','10','直教生死相許','月亮']
i=0
score=0
total=len(q_list)
while i < total:
  q = q_list[i]
  a = a_list[i]
  
  print(q,end='   ')
  ans = input().lower().strip()

  if ans == a:
    score += 10
    print('答對了! 不愧是答題高手!')
  else:
    print('請再接再厲~~~')
    print(f'答案是: {a_list[i]}')
  i += 1 #準備出下一題
  print()
print(f'總得分: {score}')
```
