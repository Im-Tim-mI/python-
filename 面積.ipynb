```python
def rect_calc(w, h):
  if (w < 0 or h < 0):
    return
  return [ ((w+h)*2,w*h), 'width==height' if w==h else 'width!=height']
```

```python
width = float(input('Width: '))
height = float(input('Hheight: '))
```

```python
(perimeter, area), rectangle = rect_calc(width, height)
```

```python
print(f'{rectangle} primeter: {perimeter}')
print(f'{rectangle} Area: {area}')
```
