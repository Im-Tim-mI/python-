```python
print('攝氏溫度轉換器')
c_temp=float(input('請輸入攝氏溫度C:'))
f_temp=(9.0*c_temp)/5.0+32.0
degree_sign=u'\N{DEGREE SIGN}'
print(f'攝氏{c_temp}{degree_sign}C等於華氏{f_temp}{degree_sign}F')
```

```python
print('華氏溫度轉換器')
f_temp=float(input('請輸入華氏溫度C:'))
c_temp=(f_temp-32.0)*5.0/9.0
degree_sign=u'\N{DEGREE SIGN}'
print(f'華氏{f_temp}{degree_sign}F等於攝氏{c_temp}{degree_sign}C')
```

```python
import math
print('圓周長 與 面積轉換器')
r=float(input('請輸入半徑:'))
PI=math.pi
f_temp=PI*r*r
c_temp=2*PI*r
print(f'半徑={r}CM\n圓面積={f_temp}CM2\n圓周長={c_temp}CM')
```
