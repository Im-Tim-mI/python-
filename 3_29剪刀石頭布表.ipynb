```python
print("剪刀石頭布?", end="")
ans=input()
if(ans == '剪刀'):
  print('你出剪刀我出布')
elif(ans == '石頭'):
  print('你出石頭我出剪刀')
elif(ans == '布'):
  print('心情起起落落像跳舞')
else:
  print('不是剪刀,不是石頭,更不是布,亂出拳')
```
